```python
import pandas as pd
import json
```

```python
with open('./aminer_papers_0/aminer_papers_0.txt') as file:
    status = []
    for line in file:
        
        status.append(json.loads(line))
        
df = pd.DataFrame(status)
df.head()
# data = pd.read_json('', lines=True)
# data.head()
```

```python
clean_df = df.drop(['pdf','url', 'volume', 'doi', 'isbn', 'issn','page_end','page_start'],axis=1)
clean_df.head()
```

```python
 clean_df.describe()
```

```python
clean_df['n_citation'].fillna(0, inplace = True)
```

```python
clean_df.head()
```

```python
clean_df['references'] = clean_df['references'].apply(lambda d: d if isinstance(d, list) else [])
dict = {}
def fill_dict(list_ref, id):
#     print(ref)
#     list_ref = json.loads(ref)
#     print(list_ref)
    for ref in list_ref:
        if ref in dict:
            dict[ref].append(id)
        else:
            dict[ref] = [id]

for index, row in clean_df.iterrows():
   fill_dict(row['references'],row['id'])
```
