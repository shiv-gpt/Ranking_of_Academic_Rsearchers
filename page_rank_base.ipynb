```python
import pandas as pd
import json
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
class Paper:
    def __init__(self):
        self.outlinks = []
        self.inlinks = []
        self.abstract = ""
        self.keywords = []
        self.page_rank = 1
        self.n_citations = None
        self.authors = []
#         self.authors_org = []
        self.journal = None
        self.year = None
        self.title = None
```

```python
dict = {}
```

```python
def init_dict(df):
    for index, row in df.iterrows():
        id = row["id"]
        if id not in dict.keys():
            dict[id] = Paper()
            dict[id].outlinks = [] if not isinstance(row["references"], list) else row["references"]
            dict[id].abstract = "" if(row["abstract"]==None) else row["abstract"]
            dict[id].keywords = [] if not isinstance(row["keywords"], list) else row["keywords"]
            dict[id].n_citations = row["n_citation"]
    #         authors = json.loads(row["authors"])
    #         print(row["authors"])
            if isinstance(row["authors"], list):
                for author in row["authors"]:
                    org = ""
                    name = ""
                    if "org" in author.keys():
                        org = author["org"]
                    if "name" in author.keys():
                        name = author["name"]

                    dict[id].authors.append((name, org))

            dict[id].journal = "" if(row["venue"]==None) else row["venue"]
            dict[id].year =  1700 if(row["year"]==None) else row["year"]
            dict[id].title = "" if(row["title"]==None) else row["title"]
        
def fill_dict():
    for id in dict.keys():
        refs = dict[id].outlinks
#         print(refs)
        for ref in refs:
            if ref in dict.keys():
                dict[ref].inlinks.append(id)
```

```python
# refer here  : http://dpk.io/pagerank
def page_rank(maxiters, epsilon):
    for i in range(maxiters):
        print(i)
        old_ranks = {}
        for id in dict.keys():
            old_ranks[id] = dict[id].page_rank
            page_rank = 0
            citations = dict[id].inlinks
            for cite_id in citations:
                page_rank += dict[cite_id].page_rank/len(dict[cite_id].outlinks)
            dict[id].page_rank = page_rank
        delta = sum([abs(old_ranks[id] - dict[id].page_rank) for id in dict.keys()])
        if delta < epsilon:
            break
```

```python
page_rank(10,0.001)
```
